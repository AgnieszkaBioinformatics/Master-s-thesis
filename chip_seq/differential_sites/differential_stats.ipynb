```python
import matplotlib.pyplot as plt

# Data
results = {'lhp1 | 3h1': {'up': 252, 'down': 120}, 
           '3h1 | WT': {'up': 168, 'down': 99}, 
           'lhp1_3h1 | 3h1': {'up': 9, 'down': 10},
           'lhp1_3h1 | lhp1': {'up': 68, 'down': 199},
           'lhp1 | WT': {'up': 8, 'down': 16},
           'lhp1_3h1 | WT': {'up': 126, 'down': 400}}

# Extract data for plotting
categories = list(results.keys())
up_values = [results[cat]['up'] for cat in categories]
down_values = [results[cat]['down'] for cat in categories]

# Create bar plot
bar_width = 0.35
index = range(len(categories))

fig, ax = plt.subplots(figsize=(8, 6))

bars1 = ax.bar(index, up_values, bar_width, label='Up', color='lightcoral')
bars2 = ax.bar([i + bar_width for i in index], down_values, bar_width, label='Down', color='royalblue')

# Add labels, title, and grid
ax.set_xlabel('Contrast')
ax.set_ylabel('Sites')
ax.set_title('Differentially bound sites')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(categories)
ax.legend()

# Add value labels on top of bars
for bars in [bars1, bars2]:
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval + 5, int(yval), ha='center', va='bottom')

# Enhance appearance
plt.tight_layout()
plt.show()
```
